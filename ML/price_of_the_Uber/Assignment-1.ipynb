```python
def haversine (lon_1, lon_2, lat_1, lat_2):
    lon_1, lon_2, lat_1, lat_2 = map(np.radians, [lon_1, lon_2, lat_1, lat_2])
    diff_lon = lon_2 - lon_1
    diff_lat = lat_2 - lat_1
    km = 2 * 6371 * np.arcsin(np.sqrt(np.sin(diff_lat/2.0)**2 + np.cos(lat_1) * np.cos(lat_2) * np.sin(diff_lon/2.0)**2))
    return km
```

```python
# Using the InterQuartile Range to fill the values
def remove_outlier(df1 , col):
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    data[col] = np.clip(df1[col] , lower_whisker , upper_whisker)
    return df1

def treat_outliers_all(df1 , col_list):
    for c in col_list:
        df1 = remove_outlier(data , c)
    return df1
```

```python
# uber_price_prediction_fixed.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

# 1. Load
data = pd.read_csv('uber.csv')
print("Initial shape:", data.shape)

# 2. Drop unnecessary cols if present
for c in ['Unnamed: 0', 'key']:
    if c in data.columns:
        data.drop(columns=c, inplace=True)

# 3. Parse datetime properly and extract features
if 'pickup_datetime' in data.columns:
    data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'], errors='coerce')
    data['month'] = data['pickup_datetime'].dt.month
    data['hour'] = data['pickup_datetime'].dt.hour
    data['day_of_week'] = data['pickup_datetime'].dt.dayofweek
    data.drop(columns=['pickup_datetime'], inplace=True)

# 4. Impute missing values column-wise (only for relevant cols)
for col in ['dropoff_longitude', 'dropoff_latitude', 'pickup_longitude', 'pickup_latitude', 'fare_amount', 'passenger_count']:
    if col in data.columns:
        data[col].fillna(data[col].median(), inplace=True)

# 5. Fix logically invalid values (per column)
if 'passenger_count' in data.columns:
    # replace non-positive passenger counts with median (or 1)
    data.loc[data['passenger_count'] <= 0, 'passenger_count'] = int(data['passenger_count'].median())

if 'fare_amount' in data.columns:
    # replace non-positive fares with median fare
    data.loc[data['fare_amount'] <= 0, 'fare_amount'] = data['fare_amount'].median()

# 6. Feature engineering: haversine distance (km)
def haversine(lon1, lon2, lat1, lat2):
    lon1, lon2, lat1, lat2 = map(np.radians, [lon1, lon2, lat1, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    km = 2 * 6371 * np.arcsin(np.sqrt(a))
    return km

coords = {'pickup_longitude','dropoff_longitude','pickup_latitude','dropoff_latitude'}
if coords.issubset(data.columns):
    data['distance_km'] = haversine(
        data['pickup_longitude'],
        data['dropoff_longitude'],
        data['pickup_latitude'],
        data['dropoff_latitude'],
    )
    # replace 0 or tiny distances with median distance (possible noisy rows)
    med_dist = data['distance_km'].median()
    data.loc[data['distance_km'] <= 0, 'distance_km'] = med_dist

# 7. Outlier treatment: clip by IQR for chosen numeric cols
def clip_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)
    return df

cols_to_clip = [c for c in ['fare_amount','pickup_longitude','pickup_latitude',
                            'dropoff_longitude','dropoff_latitude','passenger_count','distance_km'] if c in data.columns]
for c in cols_to_clip:
    data = clip_iqr(data, c)

# 8. Quick visual (optional) - uncomment if running interactively
# plt.figure(figsize=(6,4)); sns.histplot(data['fare_amount'], bins=40, kde=True); plt.title('Fare distribution'); plt.show()
# plt.figure(figsize=(6,4)); sns.scatterplot(x='distance_km', y='fare_amount', data=data, alpha=0.4); plt.show()

# 9. Correlation
corr_matrix = data.corr()
if 'fare_amount' in corr_matrix.columns:
    print("Top correlations with fare_amount:\n", corr_matrix['fare_amount'].abs().sort_values(ascending=False).head(10))
    plt.figure(figsize=(10,8))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Correlation matrix')
    plt.show()

# 10. Prepare X and y (use sensible features)
feature_candidates = ['distance_km','passenger_count','hour','month','day_of_week',
                      'pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude']
features = [f for f in feature_candidates if f in data.columns]
X = data[features].copy()
y = data['fare_amount'].copy()

print("Using features:", features)
print("X shape:", X.shape, "y shape:", y.shape)

# 11. Scale features for Linear Regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 12. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.15, random_state=42)
print("Train/test shapes:", X_train.shape, X_test.shape)

# 13. Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

mae_lr = metrics.mean_absolute_error(y_test, y_pred_lr)
mse_lr = metrics.mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = metrics.r2_score(y_test, y_pred_lr)

print("\nLinear Regression results:")
print(f"MAE: {mae_lr:.4f}")
print(f"MSE: {mse_lr:.4f}")
print(f"RMSE: {rmse_lr:.4f}")
print(f"R2: {r2_lr:.4f}")

# 14. Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

mae_rf = metrics.mean_absolute_error(y_test, y_pred_rf)
mse_rf = metrics.mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = metrics.r2_score(y_test, y_pred_rf)

print("\nRandom Forest results:")
print(f"MAE: {mae_rf:.4f}")
print(f"MSE: {mse_rf:.4f}")
print(f"RMSE: {rmse_rf:.4f}")
print(f"R2: {r2_rf:.4f}")

# 15. Compare
compare = pd.DataFrame({
    'model': ['LinearRegression','RandomForest'],
    'MAE': [mae_lr, mae_rf],
    'RMSE': [rmse_lr, rmse_rf],
    'R2': [r2_lr, r2_rf]
})
print("\nComparison:")
print(compare)

# 16. Plot Actual vs Predicted (RandomForest)
plt.figure(figsize=(6,5))
plt.scatter(y_test, y_pred_rf, alpha=0.4)
plt.xlabel('Actual fare')
plt.ylabel('Predicted fare (RF)')
plt.title('Random Forest: Actual vs Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # 45-degree
plt.show()
```
